## Load & plot the data


```python
import torch
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = torch.load("data.pt")
```

```python
# Set the global parameters for all the plots
plt.rc("axes", axisbelow=True)
plt.rc("font", **{"family": "serif", "serif": ["Cambria Math"]})
plt.rc("text", usetex=True)
```

```python
fig = plt.figure(figsize=(8, 8))

# Visuals
sns.despine(bottom=True, left=True)
ax = fig.gca()
ax.tick_params(axis="both", labelsize=20)
ax.grid(True, linestyle="--", alpha=0.7)
ax.set_xticks([-6, -3, 0, 3, 6])
ax.set_yticks([-6, -3, 0, 3, 6])
plt.xlim(-6.02, 6.02)
plt.ylim(-6.02, 6.02)
plt.xlabel("Feature X", fontsize=30)
plt.ylabel("Feature Y", fontsize=30)
ax.set_aspect("equal")

# Plot the data
sns.scatterplot(
    x=data[0, :, 0],
    y=data[0, :, 1],
    alpha=0.9,
    s=50,
    color="#c96004",
    label="Class A",
    linewidth=0.0,
    edgecolor="grey",
)
sns.scatterplot(
    x=data[1, :, 0],
    y=data[1, :, 1],
    alpha=0.9,
    s=50,
    color="#e9bf9a",
    label="Class B",
    linewidth=0.0,
    edgecolor="grey",
)


ax.legend(fontsize=24)
plt.tight_layout()
plt.savefig("data.pdf")
```
